## アヤメ= iris のデータでクラスタリング

slide12.pptx slide 23

- データの準備
- アヤメのデータを使用する
- setosaを10件、versicolorを10件、verginicaを10件抜き出す
- 質的変数のSpeciesを削除する
- 階層的クラスター分析
- 非階層的クラスター分析


```r

data(iris)
head(iris)
```

```r
#  3つの品種から10件ずつデータを抽出
iris2 <- iris[ c(1:10, 51:60, 101:110),  ]
#  質的変数のSpeciesを削除
iris3 <- iris2[ , -5 ]
```

```r
# 距離行列をもとめる


iris.dist <- dist( iris3, method = "euclidean" )
# iris.dist <- dist( iris3, method = "manhattan" )
# iris.dist <- dist( iris3, method = "maximum" )
```

```r
# 階層クラスタリングをする
# hc1 ward法 ward slide 16
# hc2 最近隣 single -> slide 12
# hc3 最遠隣 complete -> silde 13
# hc4 重心 centroid -> slide 15

res.hc1 <- hclust( iris.dist, method = "ward.D2" )
res.hc2 <- hclust( iris.dist, method = "single" )
res.hc3 <- hclust( iris.dist, method = "complete" )
res.hc4 <- hclust( iris.dist, method = "centroid" )
```

```r
#  デンドログラムの作図
plot( res.hc1, hang = -1 )
#  デンドログラムの分割
rect.hclust( res.hc1, k = 3)
```

```r
#  デンドログラムの作図
plot( res.hc2, hang = -1 )
#  デンドログラムの分割
rect.hclust( res.hc2, k = 3)
```

```r
#  デンドログラムの作図
plot( res.hc3, hang = -1 )
#  デンドログラムの分割
rect.hclust( res.hc3, k = 3)
```

```r
#  デンドログラムの作図
plot( res.hc4, hang = -1 )
#  デンドログラムの分割
rect.hclust( res.hc4, k = 3)
```
