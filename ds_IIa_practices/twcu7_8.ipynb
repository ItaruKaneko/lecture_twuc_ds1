## 1 データの読み込み
page 8

```r
#  データの呼び出し
banknote= read.table('banknote.txt', sep = ' ', header=TRUE )
head(banknote)
```

## 2 ホールドアウト検証のための bn.train, bn.test の作成
page 8

```r
#  データの分割（ホールドアウト検証）
a <- seq( 1, 200, by = 2 )
bn.train <- banknote[ a,  ]
bn.test <- banknote[ -a,  ]
head(bn.train)
head(bn.test)
```

## 3 ロジスティック回帰の実行(bn.train を使用)
page 8,9

```r
#  ロジスティック回帰の実行
res.glm1 <- glm( Status ~ Length + Bottom + Top,data = bn.train,family = binomial)
```

```r
#  結果の出力
summary( res.glm1 )
#  モデル選択
res.glm2 <- step( res.glm1 )
```

## 4 予測の試行(bn.test を使用)
page 10

### 4.1 ホールドアウト検証

```r
#  ホールドアウト検証
pred.glm1 <- predict( res.glm2, bn.test, type = "response" )　　# 確率


```

```r
pred.glm2 <- round( pred.glm1 )　　# 四捨五入
```

```r
pred.glm1
pred.glm2
```

### 4.2 実測値と予測値のデータを作成する
page 10

```r
#  実測値と予測値のデータを作成
# 以下一行はスライドのコードを修正するとうまく動きます。
tab.glm1 <- data.frame( Status = bn.test, Predict = pred.glm2 ) # <- 元のコード
# tab.glm1 <- data.frame( Status = bn.test[,1], Predict = pred.glm2 ) # <- 修正したコード
head(tab.glm1)
```

### 4.3 御判別率の計算
ここでちょっとした問題がありました。

```r
#  0, 1をカテゴリに置換（1をgenuine、0をcounterfeit）

tab.glm1[, 2] <- ifelse( tab.glm1[, 2] == 1, "genuine", "counterfeit" )
head(tab.glm1)
```

```r
#  誤判別率の計算
# 上に示した問題があると、実際にはここで実行がとまります。少しさかのぼって原因を見つける必要があります。
tab.glm2 <- table( tab.glm1 )

print(tab.glm2)
```

```r
1 - sum( diag( tab.glm2 ) ) / sum( tab.glm2 )
```
