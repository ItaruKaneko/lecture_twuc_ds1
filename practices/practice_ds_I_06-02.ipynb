## 標準偏差を求める

stdde

```r
cat('sd = ', sd(x)
```

```r
mean1 <- mean(x)
off1 <- x - mean1
print(off1)
```

```r
len1 = length(x)
rms1 <- (sum(off1**2.0)/len1) ** 0.5
rms2 <- (sum(off1**2.0)/(len1-1)) ** 0.5
sd1 <- sd(x)
cat('rms1 = ', rms1, '  rms2 = ', rms2, '  sd1 = ',sd1)
```

```r
len2 <- c(25,50,100,200)
rmsv1 <- numeric(4)
rmsv2 <- numeric(4)
for (i1 in 1:4) {
    len1 <- len2[i1]
    x1 <- x[1:len1]
    mean1 <- mean(x1)
    off1 <- x1 - mean1
    rms1[i1] <- (sum(off1**2.0)/len1) ** 0.5
    rms2[i1] <- (sum(off1**2.0)/(len1-1)) ** 0.5
}
```

```r
# プロットの初期設定
plot(rms1, type = 'l', col = 'blue', ylim = range(c(rms1, rms2)),
     xlab = 'Index', ylab = 'Values', main = 'Line Plot of rms1 and rms2')

# rms2を既存のグラフに追加
lines(rms2, col = 'red')
```

### しょうがないのでもっと大規模に


```r

for (i2 in 1:9) {
    x <- rnorm(10000,mean = 0, sd = 1)
    len2 <- c(10,100,1000,10000)
    rms1 <- numeric(4)
    rms2 <- numeric(4)
    for (i1 in 1:4) {
        len1 <- len2[i1]
        x1 <- x[1:len1]
        mean1 <- mean(x1)
        off1 <- x1 - mean1
        rms1[i1] <- (sum(off1**2.0)/len1) ** 0.5
        rms2[i1] <- (sum(off1**2.0)/(len1-1)) ** 0.5
    }
    if (i2 == 1) {
    plot(rms1, type = 'l', col = 'blue', ylim = range(c(rms1, rms2)),
        xlab = 'Index', ylab = 'Values', main = 'Line Plot of rms1 and rms2')
    }
    else {
        lines(rms1, col = 'blue')
    }
    lines(rms2, col = 'red')
}
```

```r
n_try <- 10
rms_m1 <- rep(0.0,4)
rms_m2 <- rep(0.0,4)
for (i2 in 1:n_try) {
    x <- rnorm(10000,mean = 0, sd = 1)
    len2 <- c(10,100,1000,10000)
    rms1 <- numeric(4)
    rms2 <- numeric(4)
    for (i1 in 1:4) {
        len1 <- len2[i1]
        x1 <- x[1:len1]
        mean1 <- mean(x1)
        off1 <- x1 - mean1
        rms1[i1] <- (sum(off1**2.0)/len1) ** 0.5
        rms2[i1] <- (sum(off1**2.0)/(len1-1)) ** 0.5
        rms_m1[i1] <- rms_m1[i1] + rms1[i1]
        rms_m2[i1] <- rms_m2[i1] + rms2[i1]
    }
    if (i2 == 1) {
    plot(rms1, type = 'l', col = 'blue', ylim = c(0.5,1.5),
        xlab = 'Index', ylab = 'Values', main = 'Line Plot of rms1 and rms2')
    }
    else {
        lines(rms1, col = 'blue')
    }
    lines(rms2, col = 'red')
}
lines(rms_m1/n_try, col = 'orange')
lines(rms_m2/n_try, col = 'green')
```
